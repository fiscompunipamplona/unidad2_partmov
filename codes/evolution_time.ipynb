```python
import numpy as np
import time
import matplotlib.pyplot as plt
from IPython import display
```

```python
time0 = 0.
timef = 8
timerange = np.arange(time0,timef,.25)

# integration time
xrangei = 0
xrangef = 8
xrange = np.arange(xrangei,xrangef,.5)

def position1 (t0, x0):
    y= 3*np.cos(0.785*x0 - 0.785*t0) + 4*np.cos(0.785*x0 + 0.785*t0)
    return y;
```

```python
par0_pos = []

for xx in xrange:
    par0_pos.append(position1(time0,xx))
```

```python
for tt in timerange:
    n = 0
    for xx in xrange:
        par0_pos[n] = position1(tt,xx)
        n = n+1

    plt.gca().cla()
    plt.plot(xrange,par0_pos,'b-*',label="Position")
    plt.grid(True)
    display.clear_output(wait=True)
    display.display(plt.gcf())                
    time.sleep(.5)
```
